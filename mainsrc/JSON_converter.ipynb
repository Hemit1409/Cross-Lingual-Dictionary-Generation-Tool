# Make a dictionary of the gujarati word in json file
-----
AIM: I want to create dynamic dictionary of gujarati words in which we can search the word.
it will look like this
```json
 {
     "3": {
            "હ": ['હવણ', 'હવણાં', 'હવન', 'હવસ', 'હવસી'],
            'જ': ['જુવારું', 'જુવાળ', 'જુસ્સો', 'જુહાર']
          },
     "5": {
            "હ": ['હવનપડી', 'હવસખોર', 'હવસખોરી'],
            "જ": ['જુસ્સાદાર']},
     "2": {
            "હ": ['હવા'],
            "જ": ['જુવું']
          },
     "4": {
            "જ": ['જુસ્તેજુ', 'જુસ્તેજૂ', 'જુસ્સાળું']
          }
}
```

````python
# Resources used:
matra_guj = ['ા', 'િ', 'ી', 'ુ', 'ૂ', 'ે', 'ૈ', 'ો', 'ૌ', 'ૅ', 'ૉ', 'ં', 'ૃ', '્', 'ઃ', 'ઁ', '઼', 'ઽ', 'ૄ']
matra_hin = ['ा', 'ि', 'ी', 'ु', 'ू', 'े', 'ै', 'ो', 'ौ', 'ॅ', 'ॉ', 'ं', 'ृ', '्', 'ः', 'ँ', '़', 'ऽ', 'ॄ']

gu_vow = ['અ', 'આ', 'ઇ', 'ઈ', 'ઉ', 'ઊ', 'ઋ', 'ૠ', 'ઍ', 'એ', 'ઐ', 'ઑ', 'ઓ', 'ઔ', 'ૐ']
hi_vow = ['अ', 'आ', 'इ', 'ई', 'उ', 'ऊ', 'ऋ', 'ॠ', 'ऍ', 'ए', 'ऐ', 'ऑ', 'ओ', 'औ', 'ॐ']

gu_cons = ['ક', 'ખ', 'ગ', 'ઘ', 'ઙ', 'ચ', 'છ', 'જ', 'ઝ', 'ઞ', 'ટ', 'ઠ', 'ડ', 'ઢ', 'ણ', 'ત', 'થ', 'દ', 'ધ', 'ન', 'પ', 'ફ',
           'બ', 'ભ', 'મ', 'ય', 'ર', 'લ', 'ળ', 'વ', 'શ', 'ષ', 'સ', 'હ']
hi_cons = ['क', 'ख', 'ग', 'घ', 'ङ', 'च', 'छ', 'ज', 'झ', 'ञ', 'ट', 'ठ', 'ड', 'ढ', 'ण', 'त', 'थ', 'द', 'ध', 'न', 'प', 'फ',
           'ब', 'भ', 'म', 'य', 'र', 'ल', 'ळ', 'व', 'श', 'ष', 'स', 'ह']
````

````python
# count the length of the words
def my_len(s):
    count = 0
    for i in s:
        if i not in matra_guj and i not in matra_hin:
            count += 1
    return count
````

## Adding the word of the list

````python
# readfile myListGuj.txt
import os

script_path = os.path.abspath('')  # i.e. /path/to/dir/foobar.py
script_dir = os.path.split(script_path)[0]  #i.e. /path/to/dir/
script_dir2 = os.path.split(script_dir)[0]
# print(script_dir)

my_guj_token = script_dir + "\data\myListGuj.txt"
json_container = script_dir + "\data\json_container.json"
guj_file = open(my_guj_token, "r", encoding="utf-8")

guj_content = [line.rstrip() for line in guj_file.readlines()]

# Will convert it into json file of the dictionary
guj_content_dict = {}

for i in guj_content:
    len = my_len(i)
    if len in guj_content_dict.keys():
        if(i[0] in guj_content_dict[len]):
            guj_content_dict[len][i[0]].append(i)
        else:
            guj_content_dict[len][i[0]] = [i]
    else:
        guj_content_dict[len] = {i[0]: [i]}

# print(guj_content_dict)
````

# Writing the dictionary into json file

````python
import json
#NOTE: if you open the json file you will only see the corresponding
# ascii values hence to see the actual file you have to load that file into python
with open(json_container, "w") as outfile:
    json.dump(guj_content_dict, outfile)
````
